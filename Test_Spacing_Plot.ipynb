```python
import json
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
```

```python
font = {'size'   : 20}

matplotlib.rc('font', **font)
```

$a_{c1}, n_{c1}$ randomized: Show that $\Delta d_p^*$ is always positive but not additive.

```python
with open('data/singlecoat_test_spacing.json', 'r') as f:
    data = json.load(f)
    
df = pd.DataFrame(data)

n_1 = df.n_1
a_1 = df.a_1*1e3
spacing = df.spacing
time = df.compute_time

delta_d = 1e3*(df['a_p*1'] - df['a_p*0'])*2
delta_n = 1e3*(df['n_p*1'] - df['n_p*0'])
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (13,6), sharex=True)

color = 'tab:red'
ax1.set_xlabel('Spacing [nm]')
ax1.set_ylabel('$\Delta d_p^*$ [nm]')
ax1.scatter(spacing[:3], delta_d[:3], color=color, s=70)
ax1.text(0.03, 8.45, '$n_1 = 1.45$nm', fontsize=20, bbox=dict(facecolor='white', alpha=0.5))
ax1.grid(alpha=0.3)

color = 'tab:red'
ax2.scatter(spacing[3:], delta_d[3:], color=color, s=70)
ax2.text(0, 26.9, '$n_1 = 1.70$nm', fontsize=20, bbox=dict(facecolor='white', alpha=0.5))
ax2.set_xlabel('Spacing [nm]')
ax2.grid(alpha=0.3)

fig.tight_layout(pad=1)
plt.show()

fig.savefig('figures/spacing_delta_d.png')
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (13,6), sharex=True)

color = 'tab:blue'
ax1.set_xlabel('Spacing [nm]')
ax1.set_ylabel('$\Delta n_p^*$[ppt]')
ax1.scatter(spacing[:3], delta_n[:3], color=color, s=70)
ax1.text(0.03, -0.21, '$n_1 = 1.45$nm', fontsize=20, bbox=dict(facecolor='white', alpha=0.5))
ax1.grid(alpha=0.3)

ax2.scatter(spacing[3:], delta_n[3:], color=color, s=70)
ax2.text(0, 0.46, '$n_1 = 1.70$nm', fontsize=20, bbox=dict(facecolor='white', alpha=0.5))
ax2.set_xlabel('Spacing [nm]')
ax2.grid(alpha=0.3)

fig.tight_layout(pad=1)
plt.show()

fig.savefig('figures/spacing_delta_n.png')
```

```python
fig, ax = plt.subplots(figsize=(6,5))
ax.scatter(spacing[:3], time[:3], color='tab:green', s=60)
ax.grid(alpha=0.3)
ax.set_xlabel('Spacing [nm]')
ax.set_ylabel('Computation time [s]')
fig.tight_layout(pad=1)
plt.show()
fig.savefig('figures/spacing_time.png')
```
